```python
import re
```

```python
txt = "Hi! This is Abdullah al Azmi. My age is 22. My weight is 70000 gm. My phone number is +880-165-634-2557. My room number is 124. My email is azmi@gmail.com"
```

```python
x = re.findall(r"[a-m]", txt)
print(x)
```

```python
y = re.findall(r"\d", txt)
print(y)
```

```python
z = re.findall(r"\d+", txt)
print(z)
```

```python
x = re.findall(r"\w", txt)
print(x)
```

```python
y = re.findall(r"\w+", txt)
print(y)
```

```python
z = re.findall(r"\s", txt)
print(z)
```

```python
x = re.findall(r"\s+", txt)
print(x)
```

```python
y = re.findall(r"\D", txt)  # Not digit
print(y)
```

```python
z = re.findall(r"\D+", txt)
print(z)
```

```python
x = re.findall(r"\W", txt)  # Not character
print(x)
```

```python
y = re.findall(r"\W+", txt)
print(y)
```

```python
z = re.findall(r"\S", txt)
print(z)
```

```python
x = re.findall(r"\S+", txt)
print(x)
```

```python
y = re.findall(r"\d{3}", txt)   # Exactly 3 digit
print(y)
```

```python
z = re.findall(r"\d{2,3}", txt)   # 3 digit or 2 digit. Don't use {2, 3}, use {2,3}
print(z)
```

```python
x = re.findall(r"[+]\d{6}-\d{3}-\d{4}", txt)   # Select phone number
print(x)
```

```python
y = re.findall(r"^\d+$", txt)    # Select digit at the last
print(y)
```

```python
z = re.findall(r".", txt)   # Wildcard character. Select anything except newline
print(z)
```

```python
x = re.findall(r"A..i", txt)
print(x)
```

```python
y = re.findall(r"....@.....\....", txt)
print(y)
```

```python
z = re.findall(r"....@.*\..*", txt)
print(z)
```

```python
z = re.findall(r"....@.*.com$", txt)
print(z)
```

```python
x = re.findall(r"[A-Za-z0-9]+@[A-Za-z0-9]+\..*", txt)
print(x)
```

```python
y = re.findall(r"\+\d+-\d+-\d+-\d+", txt)
print(y)
```

```python
z = re.findall(r"az.?i", txt)
print(z)
```

```python
x = re.findall(r"al|azmi", txt)
print(x)
```

```python
y = re.findall(r"is+", txt) # Capture and group
print(y)
```

```python
z = re.findall(r"(is)+", txt) # Capture and group
print(z)
```

```python
z = re.findall(r"(is)+", "isisisis") # Capture and group
print(z)
```

```python
x = re.findall(r"(\w+)@(\w+)\.*", txt)
print(x)

for _first, _second in x:
    print(_first, _second)
```

```python
y = re.findall(r"(.*)\s(\d+)", txt)
print(y)

for _, _second in y:
    print(f"My room number is: {_second}")
```

```python
z = re.search(r"\d+", txt)  # Use to find first entity
print(z)

print(z.start(), z.end(), z.group())
```

```python
x = re.split(r" ", txt)
print(x)
```

```python
y = re.sub(r"\d+", "24", txt)
print(y)
```

```python
pattern = re.compile(r"\d+")
```

```python
z = pattern.findall(txt)
print(z)
```

```python
x = pattern.search(txt)
print(x.group())
```

```python
y = pattern.sub("28", txt)
print(y)
```

```python
z = pattern.split(txt)
print(z)
```
